```python
import h5py
import numpy as np
file = h5py.File('datasets/eccv16_dataset_tvsum_google_audio_pa_pool5 copy.h5', 'w')

with h5py.File('datasets/eccv16_dataset_tvsum_clip_audio_pa_pool5 copy.h5', 'r') as f:
    keys = list(f.keys())
    print(keys)
    for key in keys:
        g = file.create_group(key)
        print(list(f[key].keys()))
        columns = list(f[key].keys())
        for column in columns:
            print(f[key][column][()])
            if column != 'features':
                g.create_dataset(column, data=f[key][column][()])

with h5py.File('datasets/eccv16_dataset_summe_google_pool5.h5', 'r') as f:
    keys = list(f.keys())
    print(keys)
    for key in keys:
        print(list(f[key].keys()))
        columns = list(f[key].keys())
        for column in columns:
            print(f[key][column][()])
            if column == 'features':
                file[key].create_dataset(column, data=f[key][column][()])
```
